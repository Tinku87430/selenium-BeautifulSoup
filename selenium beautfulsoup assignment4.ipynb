```python
import bs4
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver=webdriver.Chrome()
driver.get("https://500.co/startups?filter=1&region=&sector=&platform=")
```

```python
containers=driver.find_elements_by_xpath('//tbody[@id="portfolioContainer"]/tr')
```

```python
while True:
    try:
        WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.xpath,"//*[contains(text(),'Load More')]"))).click()
        break
    except Exception as ex:
        if 'intercepted' in str(ex):
            print(str(ex))
            pass
        else:
            break
        
```

```python
for container in containers:
    name=container.find_element_by_xpath('.//td[@class="portfolio-name"]')
    website=container.find_element_by_xpath('.//td[@class="portfolio-url"]')
    if website:
        website=website.get_attribute('href')
    country=container.find_element_by_xpath('.//td[@class="portfolio-country"]')
    try:
        sector=container.find_element_by_xpath('.//td[@class="portfolio-sector"]')
    except:
        sector=""
    platform=container.find_element_by_xpath('.//td[@class="portfolio-platform"]')
```

```python
print(name,website,country,sector,platform)
```

## Question 2 IMDB's Top rated movies

```python
url="https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc"
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.content)
```

```python
name=soup.find_all("h3",class_="lister-item-header")
```

```python
name
```

```python
movies_name=[]
for i in name:
    for j in i.find_all("a"):
        movies_name.append(j.text.replace("\n",""))
movies_name[:20]
```
